```python
# on va assigner une variable à cette liste
airbnb_cols = airbnb.select_dtypes(include=['float']).columns.to_list()
```

```python
# on va boucler avec for et une variable itératrice col sur l'objet que l'on souhaite parcourir airbnn_cols
########## faire bien attention à l'indentation (corps de la boucle) ainsi qu'aux deux point ":"

for col in airbnb_cols:
    print(col)
```

```python
# utilisation de la fonction mean pour afficher la moyenne des colonnes de l'objet airbnb_cols
for col in airbnb_cols:
    print(np.mean(airbnb[col]))
```

```python
# fonction de moyenne sur la variable calc_mean pour afficher la moyenne des colonnes du dataframe airbnb_cols

for col in airbnb_cols:
    # ft mean() appliqué à un objet series du dataframe airbnb avec la variable itératrice col
        calc_mean = np.mean(np.mean(airbnb[col]))
        print('Moyenne de la colonne ' + col + ' : ', calc_mean)
```

```python
# définition d'une fonction en python avec caractèristiques de styntaxes

# ex: on met def, puis nom de focntion, puis en paramètres un dataframe, puis le corps de la fonction
# on va split deux parties du df en deux parties égale
 
def split_function(df):
    # assignation variable df1 qui contient la moitié de df avec le param len(df/2) par la méthode loc
    df1=df.loc[0:len(df)/2]
    # assignation variable df2 qui contient la moitié de df avec le param len(df/2) après le loc sur la seconde moitié du loc
    df2=df.loc[len(df)/2:]
    return (df1, df2)

airbnb1, airbnb2 = split_function(airbnb)
# les variables dans le corps de la fonction se répètent dans les varibiables airbnb1 et airbnb2
```

```python
# on peut les observer

airbnb1.info()
```

```python
# ft pour appliquer un +100 au dataframe airbnb_cols qui contient la list des colonnes en float
# avec la méthode apply

def add_hundred(x):
    return x + 100
airbnb[airbnb_cols].apply(add_hundred)
```

```python
# pour simplifier, on peut aussi utiliser la méthode lambda pour écrire la ft en moins de  qu'avec corps e fonction et retour

airbnb[airbnb_cols].apply(lambda x: x + 100)
```
